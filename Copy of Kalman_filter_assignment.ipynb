# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

1.  Given the linear combination of a Gaussian state and Gaussian process noise:$$x^{-}_k = A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1}$$Since linear transformations of Gaussian random variables result in a Gaussian, $x_k^-$ is normally distributed.We find the expected value (mean):$$\mathbb{E}[x_k^-] = \mathbb{E}[A_{k-1}x^{+}_{k-1} + G_{k-1}w_{k-1}] = A_{k-1}\mathbb{E}[x^{+}_{k-1}] + G_{k-1}\mathbb{E}[w_{k-1}]$$Since $\mathbb{E}[w_{k-1}] = 0$ and $\mathbb{E}[x^{+}_{k-1}] = m_{k-1}$, we get:$$m_k^- = A_{k-1}m_{k-1}$$Next, we find the covariance. Because $x^{+}_{k-1}$ and $w_{k-1}$ are independent:$$\text{Var}(x_k^-) = A_{k-1}\text{Var}(x^{+}_{k-1})A_{k-1}^T + G_{k-1}\text{Var}(w_{k-1})G_{k-1}^T$$$$P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T$$

2. Using the measurement equation $y^{-}_k = H_k x^{-}_k + z_k$ where $z_k \sim \mathscr{N}(0, \Sigma_m)$:$$\mathbb{E}[y_k^-] = H_k\mathbb{E}[x_k^-] + \mathbb{E}[z_k] = H_k m_k^-$$Because $x_k^-$ and $z_k$ are independent, the variance is:$$\text{Var}(y_k^-) = H_k\text{Var}(x_k^-)H_k^T + \text{Var}(z_k) = H_k P_k^- H_k^T + \Sigma_m$$Thus, $y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m)$.

3. To form the joint Gaussian distribution of $x_k^-$ and $y_k^-$, we need their cross-covariance:$$\text{Cov}(x_k^-, y_k^-) = \mathbb{E}[(x_k^- - m_k^-)(y_k^- - H_k m_k^-)^T]$$Substitute $y_k^- = H_k x_k^- + z_k$:$$\text{Cov}(x_k^-, y_k^-) = \mathbb{E}[(x_k^- - m_k^-)(H_k(x_k^- - m_k^-) + z_k)^T]$$Since $x_k^-$ and $z_k$ are independent (cross-term expectation is zero):$$\text{Cov}(x_k^-, y_k^-) = \mathbb{E}[(x_k^- - m_k^-)(x_k^- - m_k^-)^T]H_k^T = P_k^- H_k^T$$Combining the means, variances, and cross-covariance yields the joint matrix:$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right)$$

4. and
5.  Using the standard rule for conditioning jointly Gaussian variables, the conditional distribution of $x_k^-$ given $y_k^- = y_k^{\mathrm{obs}}$ is Gaussian with:$$\text{Updated Mean } (m_k) = \mu_x + \Sigma_{xy}\Sigma_{yy}^{-1}(y_k^{\mathrm{obs}} - \mu_y)$$$$\text{Updated Covariance } (P_k) = \Sigma_{xx} - \Sigma_{xy}\Sigma_{yy}^{-1}\Sigma_{yx}$$Substituting the blocks from our joint distribution in part 3:$$m_k = m_k^- + P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} (y_k^{\mathrm{obs}} - H_k m_k^-)$$$$P_k = P_k^- - P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} H_k P_k^-$$By defining the Kalman Gain as $K_k = P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$, we can rewrite these as:$$m_k = m_k^- + K_k (y_k^{\mathrm{obs}} - H_k m_k^-)$$$$P_k = (I - K_k H_k) P_k^-$$This proves Part 4. For Part 5, these equations directly represent the expected value and variance:$$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}] = m_k$$$$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) = P_k$$

# Q. 1-D Example

Consider the scalar linear-Gaussian filter model
\begin{aligned}
x^-_k &= a\,x^+_{k-1} + w_{k-1},\qquad w_{k-1}\sim\mathscr N(0,q),\\
y^-_k &= h\,x^-_k + z_k,\qquad\;\;\;\; z_k\sim\mathscr N(0,r),
\end{aligned}
with prior $x_0\sim \mathscr{N}(m_0,P_0)$. Define $Y_{k}=\{y^{\mathrm{obs}}_1,\dots,y^{\mathrm{obs}}_{k}\}$.

Answer the following:

1. Show that
\begin{aligned}
m_k^- &= a\,m_{k-1},\\
P_k^- &= a^2 P_{k-1} + q.
\end{aligned}

2. Show that
\begin{aligned}
m_k &= m_k^- + K_k\,v_k
= m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^-
= \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

3. Show that the predictive measurement distribution (before seeing $y_k$):
\begin{align*}
p(y^-_k | Y_{k-1})=\mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}

4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$):
\begin{align*}
p(y^-_k\mid Y_k)=\mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

5. For suitable numerical values for $a,q,h,r$ animate the above to prior distribution and the posterior distribution.

1, 2, 3 & 4. Scalar Reduction
For a 1-D model, matrices become scalars: $A = a$, $G = 1$, $H = h$, $\Sigma_p = q$, and $\Sigma_m = r$.
Substituting these directly into our general analytical derivation above yields the required proofs:

$m_k^- = a m_{k-1}$ and $P_k^- = a^2 P_{k-1} + (1)q(1) = a^2 P_{k-1} + q$.

Let $S_k = h P_k^- h + r = h^2 P_k^- + r$. Then $K_k = \frac{P_k^- h}{S_k}$. The mean update is exactly $m_k^- + K_k(y^{\mathrm{obs}}_k - h m_k^-)$ and variance is $(1 - K_k h)P_k^- = (1 - \frac{P_k^- h^2}{S_k})P_k^-$.

The predictive measurement (prior) marginalizes out the state before observation: $p(y^-_k \mid Y_{k-1})$ has mean $h m_k^-$ and variance $h^2 P_k^- + r$, perfectly matching $\mathscr{N}(h m_k^-, h^2 P_k^- + r)$.

The posterior predictive uses the updated state estimate $m_k$ and updated variance $P_k$: $p(y^-_k \mid Y_k) = \mathscr{N}(h m_k, h^2 P_k + r)$.

5. Python Animation

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.stats import norm
from IPython.display import HTML

# Set consistent seed for simulation
np.random.seed(250)

# Parameters
a = 1.0     # State transition
h = 1.0     # Measurement scale
q = 0.5     # Process noise variance
r = 1.0     # Measurement noise variance

# Initial conditions
m_prev = 0.0
P_prev = 1.0
true_x = 0.0

frames_data = []

# Simulate 20 steps
for k in range(20):
    # True state and measurement simulation
    true_x = a * true_x + np.random.normal(0, np.sqrt(q))
    y_obs = h * true_x + np.random.normal(0, np.sqrt(r))

    # Prediction (Prior)
    m_prior = a * m_prev
    P_prior = a**2 * P_prev + q

    # Update (Posterior)
    S = h**2 * P_prior + r
    K = P_prior * h / S
    m_post = m_prior + K * (y_obs - h * m_prior)
    P_post = (1 - K * h) * P_prior

    frames_data.append((m_prior, P_prior, m_post, P_post, y_obs))

    m_prev = m_post
    P_prev = P_post

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
x_axis = np.linspace(-10, 10, 500)
line_prior, = ax.plot([], [], 'b--', label='Prior $x^-$')
line_post, = ax.plot([], [], 'r-', label='Posterior $x^+$')
obs_line = ax.axvline(0, color='g', linestyle=':', label='Observation $y/h$')

ax.set_xlim(-10, 10)
ax.set_ylim(0, 1.2)
ax.set_title('1-D Kalman Filter: Prior vs Posterior')
ax.legend(loc='upper right')

def update(frame):
    m_prior, P_prior, m_post, P_post, y_obs = frames_data[frame]

    line_prior.set_data(x_axis, norm.pdf(x_axis, m_prior, np.sqrt(P_prior)))
    line_post.set_data(x_axis, norm.pdf(x_axis, m_post, np.sqrt(P_post)))
    obs_line.set_xdata([y_obs/h, y_obs/h])

    return line_prior, line_post, obs_line

anim = FuncAnimation(fig, update, frames=20, blit=True, interval=500)
plt.close()

# Display in Colab
HTML(anim.to_html5_video())
```

# Q. 2D-Position Estimation

## Part A

Consider a two-dimensional constant-velocity dynamical system. The hidden state at time step $k$ is

$$
x_k =
\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix},
$$

where $(p_x(k),p_y(k))$ denote the position components and $(v_x(k),v_y(k))$ denote the velocity components.

The measurement consists only of the two position components:

$$
y_k =
\begin{bmatrix}
p_x^{\mathrm{meas}}(k)\\
p_y^{\mathrm{meas}}(k)
\end{bmatrix}.
$$

The linear Gaussian state-space filter model is

$$
x^-_k = A x^+_{k-1} + G w_{k-1},
$$

$$
y^-_k = Hx^+_k + z_k,
$$

where

$$
w_{k-1} \sim \mathscr{N}(0,\Sigma_p),
\qquad
z_k \sim \mathscr{N}(0,\Sigma_m).
$$

The process noise sequence $w_k$, measurement noise sequence $z_k$, and the initial state are assumed mutually independent.

Show that:

$$
A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix},
$$



$$
H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix},
$$

and

$$
G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}.
$$

## Part B

Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

Part A: State-Space DerivationFrom basic Newtonian kinematics, assuming a constant velocity model perturbed by a random acceleration (process noise) $w = [w_x, w_y]^T$, the position and velocity at step $k$ are:$$p_x(k) = p_x(k-1) + v_x(k-1)\Delta t + \frac{1}{2} w_x \Delta t^2$$$$p_y(k) = p_y(k-1) + v_y(k-1)\Delta t + \frac{1}{2} w_y \Delta t^2$$$$v_x(k) = v_x(k-1) + w_x \Delta t$$$$v_y(k) = v_y(k-1) + w_y \Delta t$$

Putting this into matrix form $x_k = A x_{k-1} + G w_{k-1}$:$$\begin{bmatrix} p_x(k) \\ p_y(k) \\ v_x(k) \\ v_y(k) \end{bmatrix} =
\begin{bmatrix} 1 & 0 & \Delta t & 0 \\ 0 & 1 & 0 & \Delta t \\ 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix}
\begin{bmatrix} p_x(k-1) \\ p_y(k-1) \\ v_x(k-1) \\ v_y(k-1) \end{bmatrix} +
\begin{bmatrix} \frac{1}{2}\Delta t^2 & 0 \\ 0 & \frac{1}{2}\Delta t^2 \\ \Delta t & 0 \\ 0 & \Delta t \end{bmatrix}
\begin{bmatrix} w_x \\ w_y \end{bmatrix}$$

This confirms the state transition matrix $A$ and the noise gain matrix $G$.Since we only measure position components $p_x$ and $p_y$, the measurement equation $y_k = H x_k + z_k$ isolates the first two state variables:$$\begin{bmatrix} p_x^{\mathrm{meas}}(k) \\ p_y^{\mathrm{meas}}(k) \end{bmatrix} =
\begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \end{bmatrix}
\begin{bmatrix} p_x(k) \\ p_y(k) \\ v_x(k) \\ v_y(k) \end{bmatrix} + z_k$$

This confirms the observation matrix $H$.



Part B: Python implementation for filtering GPS measurements

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
np.random.seed(250)
dt = 1.0  # Time step
steps = 50

# Matrices
A = np.array([
    [1, 0, dt, 0],
    [0, 1, 0, dt],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

H = np.array([
    [1, 0, 0, 0],
    [0, 1, 0, 0]
])

G = np.array([
    [0.5 * dt**2, 0],
    [0, 0.5 * dt**2],
    [dt, 0],
    [0, dt]
])

# Noise covariances
sigma_p_val = 0.1  # Process noise variance (acceleration)
sigma_m_val = 5.0  # Measurement noise variance (GPS inaccuracy)
Sigma_p = np.eye(2) * sigma_p_val
Sigma_m = np.eye(2) * sigma_m_val
Q = G @ Sigma_p @ G.T  # Effective process noise covariance in state space

# Simulate True Path and Noisy Measurements
true_states = []
measurements = []
x_true = np.array([0, 0, 1.5, 1.0])  # Start at origin, moving diagonally

for _ in range(steps):
    # True state progression
    w = np.random.multivariate_normal([0, 0], Sigma_p)
    x_true = A @ x_true + G @ w
    true_states.append(x_true)

    # Noisy GPS measurement
    z = np.random.multivariate_normal([0, 0], Sigma_m)
    y = H @ x_true + z
    measurements.append(y)

true_states = np.array(true_states)
measurements = np.array(measurements)

# 2D Kalman Filter Implementation
m = np.array([0, 0, 0, 0])  # Initial guess
P = np.eye(4) * 10.0        # Initial uncertainty
filtered_states = []

for y in measurements:
    # 1. Prediction (Prior)
    m_prior = A @ m
    P_prior = A @ P @ A.T + Q

    # 2. Update (Posterior)
    S = H @ P_prior @ H.T + Sigma_m
    K = P_prior @ H.T @ np.linalg.inv(S)

    m = m_prior + K @ (y - H @ m_prior)
    P = (np.eye(4) - K @ H) @ P_prior

    filtered_states.append(m)

filtered_states = np.array(filtered_states)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(true_states[:, 0], true_states[:, 1], 'k--', label='True Path')
plt.scatter(measurements[:, 0], measurements[:, 1], c='r', marker='x', label='Noisy GPS Measurements', alpha=0.6)
plt.plot(filtered_states[:, 0], filtered_states[:, 1], 'b-', linewidth=2, label='Kalman Filtered Path')

plt.title('2D Position Estimation using Kalman Filter')
plt.xlabel('Position X')
plt.ylabel('Position Y')
plt.legend()
plt.grid(True)
plt.show()
```
