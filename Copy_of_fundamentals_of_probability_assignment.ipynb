# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


The Sample Space ($\Omega$): $$\Omega = \{(x, y) \in \mathbb{R}^2 : x^2 + y^2 \leq 10^2\}$$

The Event ($A$): $$A = \{(x, y) \in \Omega : 9 \leq \sqrt{x^2 + y^2} \leq 10\}$$


Total Area: $\text{Area}_{\text{total}} = \pi \times 10^2 = 100\pi$

Area of Event A:$$\text{Area}_A = \pi(10^2) - \pi(9^2) = 100\pi - 81\pi = 19\pi$$
Probability $P(A)$:$$P(A) = \frac{\text{Area}_A}{\text{Area}_{\text{total}}} = \frac{19\pi}{100\pi} = 0.19$$

2. Uniform Distribution of Distance and Direction

Sample Space for Distance: The interval $[0, 10]$ has a length of $10$.

Interval for Event A: Since the point must be within $1$ unit of the boundary, the distance $r$ must be in the interval $[9, 10]$. The length of this interval is $1$.

Probability $P(A)$:
$$P(A) = \frac{\text{Length of Event A interval}}{\text{Total Length of distance interval}} = \frac{10 - 9}{10} = \frac{1}{10} = 0.1$$

3. Why the answers differ

The two answers differ because the phrase “a point is chosen at random from a disk” does not determine a unique probability model.
### Model 1:
In the first model, probability is distributed uniformly over area. Since the outer part of the disk occupies a relatively large amount of area, points near the boundary are more likely than points near the center. This gives
              $$P(A) = \frac{19}{100}$$


### Model 2: Uniform Distribution over the Radius

In the second model, probability is distributed uniformly over the **radius**. That means each radial interval of the same length is equally likely, regardless of how much area it contains.

Since the interval $[9, 10]$ has length $1$ out of a total radial length $10$, this gives:

$$P(A) = \frac{10 - 9}{10} = \frac{1}{10}$$

So the difference comes from using two different probability measures on the same geometric set. The event
 is the same in both cases, but the meaning of “chosen at random” is different.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

In a circle of radius R, the side length of an inscribed equilateral triangle is L = R$\sqrt{3}$. Any chord that has a larger length will have its midpoint closer to the center than the midpoint of the side of the inscribed triangle, which is a distance of R/2 from the center.
Here are three natural ways to select chords:

1. Random Endpoints (The "Angle" Method)

Pick a point on the circumference to be the first endpoint of the chord. Then, pick a second point uniformly at random along the circumference.The Logic: If we rotate the circle so the first point is at one vertex of the inscribed triangle, the chord will be longer than the side length only if the second point falls on the arc between the other two vertices.

The Math: This arc covers exactly $1/3$ of the circumference.Result: $P = 1/3$.

2. Random Radial Distance (The "Radius" Method)

Pick a radius of the circle, and then pick a point uniformly at random along that radius to serve as the midpoint of the chord.The Logic: This is similar to your "Model 2" in the target problem. A chord is longer than the triangle's side if its midpoint is less than $R/2$ from the center.

The Math: The favorable interval is $[0, R/2]$ within the total radial length $[0, R]$.Result: $P = 1/2$.

3. Random Midpoint (The "Area" Method)
Pick a point anywhere inside the circle uniformly at random to be the midpoint of the chord.The Logic: A point uniquely defines a chord. For the chord to be longer than the side length, this midpoint must fall within a smaller, concentric circle with radius $R/2$.The Math: This uses the "Area" logic. The area of the inner circle is $\pi(R/2)^2 = \pi R^2/4$, and the total area is $\pi R^2$.Result: $P = 1/4$.

4.Show that  σ(X)⊆σ(Y) , and explain why  Y  contains more information than  X .

To show that $\sigma(X) \subseteq \sigma(Y)$, we compare the events that each random variable can distinguish.The $\sigma$-algebra generated by $X$ is:$$\sigma(X) = \{\emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\}\}.$$The $\sigma$-algebra generated by $Y$ is the power set of the sample space:$$\sigma(Y) = \mathcal{P}(\Omega) = \{\emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \dots, \Omega\}.$$Every set in $\sigma(X)$ is also a set in $\sigma(Y)$. Specifically:$\{G_1, G_2\} = \{G_1\} \cup \{G_2\} \in \sigma(Y)$$\{D_1, D_2\} = \{D_1\} \cup \{D_2\} \in \sigma(Y)$Therefore,$$\sigma(X) \subseteq \sigma(Y).$$

In probability theory, the size of a $\sigma$-algebra represents the "resolution" of our knowledge.$X$ is a coarse measurement: It only tells you the quality of the item. Knowing $X=0$ tells you the outcome is in $\{G_1, G_2\}$, but you cannot distinguish which machine produced it.$Y$ is a fine measurement: It tells you both quality and the specific machine. Knowing $Y=1$ tells you exactly that the outcome is $G_1$.Because $X$ is a function of $Y$ ($X = f(Y)$), knowing the value of $Y$ always allows you to determine $X$. However, knowing $X$ is not sufficient to determine $Y$. Thus, $Y$ is strictly more informative than $X$.

13. Compute the differential entropy of $X$The random variable $X$ represents the temperature deviation and follows a standard normal distribution, $X \sim N(0,1)$. The differential entropy for a normal distribution $N(\mu, \sigma^2)$ is given by the formula $h(X) = \frac{1}{2}\ln(2\pi e \sigma^2)$.Substituting $\sigma^2 = 1$:$$h(X) = \frac{1}{2}\ln(2\pi e)$$Numerical Value:$$h(X) \approx 1.4189 \text{ nats}$$

14. Compute the differential entropy of $Y$The random variable $Y = (T, P)$ represents the full sensor state, where $T$ and $P$ are independent normal components with variances $\sigma_T^2 = 1$ and $\sigma_P^2 = 4$. For independent variables, the joint differential entropy is the sum of their individual entropies:$$h(Y) = h(T) + h(P)$$Using the entropy formula for both:$h(T) = \frac{1}{2}\ln(2\pi e)$$h(P) = \frac{1}{2}\ln(2\pi e \cdot 4) = \frac{1}{2}\ln(8\pi e)$Calculation:$$h(Y) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(8\pi e) = \ln(2\pi e) + \ln 2$$Numerical Value:$$h(Y) \approx 3.5310 \text{ nats}$$

15. Compute the conditional differential entropy $h(Y|X)$The conditional differential entropy $h(Y|X)$ measures the uncertainty remaining in the full sensor state $Y$ once the temperature $X$ is known. Since $T$ and $P$ are independent, knowing $T$ provides no information about $P$. Therefore, the remaining uncertainty in $Y$ is exactly the uncertainty in $P$:$$h(Y|X) = h(P)$$Calculation:$$h(Y|X) = \frac{1}{2}\ln(2\pi e \cdot 4) = \frac{1}{2}\ln(8\pi e)$$Numerical Value:$$h(Y|X) \approx 2.1121 \text{ nats}$$Physical Interpretation:Once the temperature deviation is measured and known, the only remaining uncertainty in the system is the pressure deviation.
