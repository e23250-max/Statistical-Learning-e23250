## Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)

**1. Bias of the Maximum Likelihood Estimator (MLE) and Bessel's Correction**

To derive why the raw Maximum Likelihood Estimator for the covariance matrix, $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$, is inherently biased, we examine its expectation. The MLE for the covariance matrix is defined as:
$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Taking the mathematical expectation:
$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \mathbb{E} \left[ \sum_{i=1}^{n} ((\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}))((\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}))^T \right]$$
Expanding and utilizing the fact that $\mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T] = \frac{1}{n}\boldsymbol{\Sigma}$, the cross terms simplify, yielding:
$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n-1}{n}\boldsymbol{\Sigma}$$
This demonstrates that $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$ underestimates the true population covariance.

To correct this bias, we apply Bessel's correction, multiplying the estimator by the factor $\frac{n}{n-1}$:
$$\mathbf{S} = \frac{n}{n-1} \widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n-1} \sum_{i=1}^{n} (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$
Taking the expectation proves it is an unbiased estimator:
$$\mathbb{E}[\mathbf{S}] = \frac{n}{n-1} \left( \frac{n-1}{n}\boldsymbol{\Sigma} \right) = \boldsymbol{\Sigma}$$

**2. Hypothesis Testing in Structural Health Monitoring**

* **Type I Error ($\alpha$):** Within continuous structural health monitoring, a Type I error represents a "false alarm." It occurs when the diagnostic layer rejects the null hypothesis $H_0$ (the status quo), concluding the asset is damaged or anomalous when it actually remains completely healthy.
* **Type II Error ($\beta$):** This is a "missed detection." It occurs when the diagnostic system fails to reject $H_0$, concluding the structure is healthy when a true structural failure or anomaly is present.
* **Consequence of Ultra-Conservative $\alpha$:** If an engineer sets an ultra-conservative significance threshold like $\alpha = 0.0001$ to prevent industrial plant shutdowns, the mathematical consequence is a severe reduction in the statistical power ($1-\beta$) of the diagnostic layer.
* **Geometric Modification:** By minimizing the false alarm rate, the geometric volume of the "healthy operation" confidence ellipsoid is drastically expanded, making the system blind to subtle structural drifts.

**1. Slutsky's Theorem**
Slutsky's Theorem states that if a sequence of random vectors $\mathbf{X}_n$ converges in distribution to a random vector $\mathbf{X}$ ($\mathbf{X}_n \xrightarrow{d} \mathbf{X}$), and another sequence of random matrices $\mathbf{Y}_n$ converges in probability to a constant matrix $\mathbf{C}$ ($\mathbf{Y}_n \xrightarrow{p} \mathbf{C}$), then:
$$\mathbf{Y}_n \mathbf{X}_n \xrightarrow{d} \mathbf{C}\mathbf{X}$$
$$\mathbf{X}_n + \mathbf{Y}_n \xrightarrow{d} \mathbf{X} + \mathbf{C}$$

**2. Justification for Empirical Substitution**
By the Weak Law of Large Numbers, the empirical sample covariance matrix converges in probability to the latent true population matrix as the sample size increases: $\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}$. Therefore, $\mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$.
Given the Central Limit Theorem establishes $\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$, Slutsky's Theorem allows us to substitute the latent $\boldsymbol{\Sigma}$ with the observable $\mathbf{S}$ as $n \to \infty$, operationalizing the practical parametric modeling distribution:
$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left({\boldsymbol{\mu}}, \frac{1}{n}\mathbf{S}\right)$$

## Part A: Theoretical Fundamentals (Coordinate Projections & Orthogonality)

**1. Covariance of Transformed Space and Geometric Meaning**
Given $\mathbf{Z}_i = \mathbf{P}^T \widetilde{\mathbf{X}}_i$ and $\boldsymbol{\Sigma} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^T$:
$$\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbb{E}[\mathbf{P}^T \widetilde{\mathbf{X}}_i (\mathbf{P}^T \widetilde{\mathbf{X}}_i)^T]$$
$$\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbf{P}^T \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T] \mathbf{P} = \mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}$$
Substituting the spectral decomposition:
$$\mathbf{P}^T (\mathbf{P} \mathbf{\Lambda} \mathbf{P}^T) \mathbf{P} = (\mathbf{P}^T \mathbf{P}) \mathbf{\Lambda} (\mathbf{P}^T \mathbf{P}) = \mathbf{I} \mathbf{\Lambda} \mathbf{I} = \mathbf{\Lambda}$$

*Geometric and Statistical Meaning:* The resulting strictly diagonal matrix $\mathbf{\Lambda}$ indicates that in the principal component space, all off-diagonal elements (cross-covariances between distinct coordinates $Z_{i,j}$ and $Z_{i,k}$ for $j \neq k$) are precisely zero. Geometrically, the transformed axes are perfectly orthogonal, yielding statistically independent (de-correlated) features.

**2. Variance Preservation and explained ratios**
Using the cyclic permutation property of the trace operator ($\text{tr}(\mathbf{A}\mathbf{B}\mathbf{C}) = \text{tr}(\mathbf{C}\mathbf{A}\mathbf{B})$):
$$\text{tr}(\boldsymbol{\Sigma}) = \text{tr}(\mathbf{P} \mathbf{\Lambda} \mathbf{P}^T) = \text{tr}(\mathbf{P}^T \mathbf{P} \mathbf{\Lambda}) = \text{tr}(\mathbf{I} \mathbf{\Lambda}) = \text{tr}(\mathbf{\Lambda}) = \sum_{j=1}^m \lambda_j$$
This formally proves total system variance is preserved during the mapping.
* Cumulative Explained Variance Ratio: $\Phi(k) = \frac{\sum_{j=1}^k \lambda_j}{\sum_{j=1}^m \lambda_j}$
* Residual Unexplained Variance Ratio: $\Psi(k) = 1 - \Phi(k) = \frac{\sum_{j=k+1}^m \lambda_j}{\sum_{j=1}^m \lambda_j}$

**3. Vector Reconstruction and Diagnostic Statistics**
* Reconstructed vector: $\widehat{\mathbf{x}}_i = \widehat{\mathbf{P}}_k \mathbf{z}_{i,k}$
* Residual error vector: $\mathbf{e}_i = \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$
Since $\widehat{\mathbf{P}}_k$ and $\widehat{\mathbf{P}}_{m-k}$ are orthogonal subspaces, taking the squared Euclidean norm yields the Pythagorean identity:
$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2$$

*Diagnostic Matching:* An environmental load anomaly (extreme wind loads) scales the amplitude of existing normal structural behaviors, which is captured by the primary variance pathways. Therefore, it maps to the **Hotelling's $T^2$ statistic**.
An internal structural fracture (fatigue crack) changes the physical load pathways, breaking the established correlation rules between sensors. This novel behavior violates the principal space, dropping directly into the residual subspace monitored by the **$Q$ statistic (SPE)**.

Complete the Python class method below inside Google Colab. The function must:

1. Standardize and center the incoming realizations.
2. Decompose the unbiased sample covariance matrix $\mathbf{S}$ using `np.linalg.eigh` and sort components in descending order ($\widehat{\lambda}_1 \ge \dots \ge \widehat{\lambda}_m$).
3. Loop through every possible subspace cutoff value $k \in \{1, 2, 3\}$, calculating the sample vector array for Hotelling’s $T^2$ ($T^2_i = \sum_{j=1}^k \frac{z_{i,j}^2}{\widehat{\lambda}_j}$) and the $Q$ statistic ($Q_i = \sum_{j=k+1}^m z_{i,j}^2$).

````python
import numpy as np

class StructuralDiagnostics:
    def __init__(self, data: np.ndarray):
        """
        Initializes the pipeline with a baseline archive.
        data shape should be (n_samples, m_sensors), e.g., (3000, 4)
        """
        self.X = data
        self.n, self.m = data.shape

    def analyze_truncation_boundaries(self):
        """
        Analyzes the response of T^2 and Q statistics across different
        subspace truncation boundaries (k).
        """
        # 1. Standardize and center the incoming realizations
        mu = np.mean(self.X, axis=0)
        std_dev = np.std(self.X, axis=0, ddof=1)
        X_std = (self.X - mu) / std_dev

        # 2. Compute unbiased sample covariance matrix S
        S = np.cov(X_std, rowvar=False)

        # Decompose using np.linalg.eigh
        eigenvalues, eigenvectors = np.linalg.eigh(S)

        # Sort components in descending order (eigh returns ascending by default)
        sorted_indices = np.argsort(eigenvalues)[::-1]
        self.lambdas = eigenvalues[sorted_indices]
        self.P = eigenvectors[:, sorted_indices]

        # Project data into the principal component space (Z = X * P)
        Z = X_std @ self.P

        results = {}

        # 3. Loop through every possible subspace cutoff value k ∈ {1, 2, 3}
        for k in [1, 2, 3]:
            # Calculate Hotelling's T^2 for each sample i
            # Z[:, :k] gets the first k components for all samples
            T2_array = np.sum((Z[:, :k]**2) / self.lambdas[:k], axis=1)

            # Calculate Q statistic for each sample i
            # Z[:, k:] gets the remaining (m-k) components
            Q_array = np.sum(Z[:, k:]**2, axis=1)

            # 4. Compute the expected value (empirical mean) across all samples
            mean_T2 = np.mean(T2_array)
            mean_Q = np.mean(Q_array)

            # 5. Store metrics
            results[f'k={k}'] = {
                'Mean_T2': mean_T2,
                'Mean_Q': mean_Q
            }

        return results

# --- Example Execution ---
# Generate synthetic baseline data (n=3000, m=4)
np.random.seed(42)
baseline_archive = np.random.multivariate_normal(
    mean=[0, 0, 0, 0],
    cov=[[1.0, 0.8, 0.6, 0.2],
         [0.8, 1.0, 0.7, 0.3],
         [0.6, 0.7, 1.0, 0.4],
         [0.2, 0.3, 0.4, 1.0]],
    size=3000
)

# Initialize and run the pipeline
pipeline = StructuralDiagnostics(baseline_archive)
metrics = pipeline.analyze_truncation_boundaries()

for k, vals in metrics.items():
    print(f"Subspace Boundary {k}:")
    print(f"  -> Expected T^2: {vals['Mean_T2']:.4f}")
    print(f"  -> Expected Q:   {vals['Mean_Q']:.4f}\n")
````
