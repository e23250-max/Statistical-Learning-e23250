#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

 Probability Measure $P$: Assigned to the elementary outcomes as given:$$P(\{\omega_1\}) = 0.50, \quad P(\{\omega_2\}) = 0.10, \quad P(\{\omega_3\}) = 0.20, \quad P(\{\omega_4\}) = 0.20$$

 The $\sigma$-algebra $\sigma(Y)$ is the smallest $\sigma$-algebra containing these preimages:$$\sigma(Y) = \{\emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega\}$$


3. By definition, the conditional expectation $E[X \mid Y]$ is a random variable that is $\sigma(Y)$-measurable. This means it cannot distinguish between individual outcomes within the atoms (the smallest non-empty building blocks) of $\sigma(Y)$.Because the atoms of $\sigma(Y)$ are exactly $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, $E[X \mid Y]$ must map all elements within each atom to the same value. Therefore, it must be constant on $\{\omega_1, \omega_2\}$ and constant on $\{\omega_3, \omega_4\}$.

We must check that $\int_A E[X \mid Y] \, dP = \int_A X \, dP$ for all 4 sets in $\sigma(Y)$:For $A = \emptyset$: Both sides are trivially $0$.For $A = \{\omega_1, \omega_2\}$:$\int_A E[X \mid Y] \, dP = \frac{1}{6} \cdot P(\{\omega_1, \omega_2\}) = \frac{1}{6} \cdot 0.60 = 0.10$$\int_A X \, dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0 + 1 \cdot 0.10 = 0.10$ (Matches)For $A = \{\omega_3, \omega_4\}$:$\int_A E[X \mid Y] \, dP = \frac{1}{2} \cdot P(\{\omega_3, \omega_4\}) = \frac{1}{2} \cdot 0.40 = 0.20$$\int_A X \, dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0 + 1 \cdot 0.20 = 0.20$ (Matches)For $A = \Omega$:$\int_\Omega E[X \mid Y] \, dP = 0.10 + 0.20 = 0.30$$\int_\Omega X \, dP = 0.10 + 0.20 = 0.30$ (Matches)

* $p_{X|Y}(1 \mid 0) = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{0.10}{0.60} = \frac{1}{6}$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

### Proof of Conditional Density of the Multivariate Gaussian

To find the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we exploit the properties of jointly Gaussian random vectors.

**Step 1: Define an uncorrelated transformation**
We define a new random vector $\mathbf{U}$ through a linear transformation designed to be uncorrelated with $\mathbf{Y}$:
$$\mathbf{U} = \mathbf{X} - \mathbf{A}\mathbf{Y}$$
We want to find a matrix $\mathbf{A}$ such that $\text{Cov}(\mathbf{U}, \mathbf{Y}) = \mathbf{0}$.
$$\text{Cov}(\mathbf{U}, \mathbf{Y}) = \text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{Y}) = \text{Cov}(\mathbf{X}, \mathbf{Y}) - \mathbf{A}\text{Cov}(\mathbf{Y}, \mathbf{Y})$$
$$\text{Cov}(\mathbf{U}, \mathbf{Y}) = \boldsymbol{\Sigma}_{XY} - \mathbf{A}\boldsymbol{\Sigma}_{YY}$$
Setting this to $\mathbf{0}$ gives:
$$\mathbf{A} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$$

**Step 2: Independence**
Because $\mathbf{Z} = [\mathbf{X}, \mathbf{Y}]^T$ is jointly Gaussian, any linear combination is also Gaussian. Since $\mathbf{U}$ and $\mathbf{Y}$ are uncorrelated and jointly Gaussian, they are **strictly independent**.
Therefore, conditioning on $\mathbf{Y}$ has no effect on $\mathbf{U}$. The conditional distribution of $\mathbf{U} \mid \mathbf{Y}=\mathbf{y}$ is simply the marginal distribution of $\mathbf{U}$.

**Step 3: Mean and Covariance of $\mathbf{U}$**
* Mean: $E[\mathbf{U}] = E[\mathbf{X}] - \mathbf{A}E[\mathbf{Y}] = \boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\mu}_Y$
* Covariance: $\text{Var}(\mathbf{U}) = \text{Var}(\mathbf{X} - \mathbf{A}\mathbf{Y}) = \boldsymbol{\Sigma}_{XX} - \mathbf{A}\boldsymbol{\Sigma}_{YX} - \boldsymbol{\Sigma}_{XY}\mathbf{A}^T + \mathbf{A}\boldsymbol{\Sigma}_{YY}\mathbf{A}^T$
Substitute $\mathbf{A} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$ and simplify:
$$\text{Var}(\mathbf{U}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} = \boldsymbol{\Sigma}_{X|Y}$$

**Step 4: Shift back to $\mathbf{X}$**
We can rewrite $\mathbf{X}$ in terms of $\mathbf{U}$ and $\mathbf{Y}$:
$$\mathbf{X} = \mathbf{U} + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{Y}$$
Conditioning on $\mathbf{Y} = \mathbf{y}$ treats $\mathbf{y}$ as a deterministic constant:
$$\mathbf{X} \mid (\mathbf{Y}=\mathbf{y}) = \mathbf{U} + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{y}$$

Since $\mathbf{U}$ is independent of $\mathbf{Y}$, we can calculate the conditional mean $\boldsymbol{\mu}^*$:
$$\boldsymbol{\mu}^* = E[\mathbf{X} \mid \mathbf{Y}=\mathbf{y}] = E[\mathbf{U}] + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{y}$$
$$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\mu}_Y + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{y} = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$

The conditional variance relies solely on the variance of $\mathbf{U}$:
$$\text{Var}(\mathbf{X} \mid \mathbf{Y}=\mathbf{y}) = \text{Var}(\mathbf{U}) = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} = \boldsymbol{\Sigma}_{X|Y}$$

**Step 5: Final Conditional Density Formulation**
Since linear combinations of Gaussians remain Gaussian, the conditional distribution $\mathbf{X} \mid \mathbf{Y}=\mathbf{y}$ is a multivariate normal $\mathcal{N}(\boldsymbol{\mu}^*, \boldsymbol{\Sigma}_{X|Y})$. Inserting this mean and variance into the standard multivariate normal density formula yields:
$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$
*(Note: $\mathbf{S}_{XX}$ in the prompt equation corresponds to $\boldsymbol{\Sigma}_{X|Y}$.)*

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



### 1. Joint Distribution
Since $X \sim \mathscr{N}(0, 1)$ and $Q \sim \mathscr{N}(0, 4)$ are independent:
* $E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$.
* $\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$. Thus, **$Y \sim \mathscr{N}(0, 5)$**.
* $\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q) = \text{Var}(X) + 0 = 1$.
The vector $(X, Y)$ follows a bivariate normal distribution.

### 2. Information Geometry
$X$ is **not** $\sigma(Y)$-measurable because there is no deterministic mapping (one-to-one function) from $Y$ to $X$. A single observation $Y = 5$ corresponds to an infinite line of possibilities $X + Q = 5$. The engineer cannot definitively untangle whether the high power reading came from high temperature ($X$) or a heavy mechanical load ($Q$).

### 3. Conditional Expectation
* Formula: $E[X \mid Y=y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0) = \frac{y}{5}$.
* Therefore, the estimator is $\widehat{X} = E[X \mid Y] = \frac{1}{5}Y$.
* The multiplier $\frac{1}{5}$ exactly equals the **signal-to-total-variance ratio**: $\frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} = \frac{1}{5}$.

### 4. Verification of the Residual
Let $Z = X - \widehat{X} = X - \frac{1}{5}Y = X - \frac{1}{5}(X + Q) = \frac{4}{5}X - \frac{1}{5}Q$.
* $E[Z] = \frac{4}{5}E[X] - \frac{1}{5}E[Q] = 0$.
* Orthogonality Principle: $E[ZY] = E[(\frac{4}{5}X - \frac{1}{5}Q)(X + Q)] = \frac{4}{5}E[X^2] + \frac{4}{5}E[XQ] - \frac{1}{5}E[XQ] - \frac{1}{5}E[Q^2]$.
Since $X$ and $Q$ are independent, $E[XQ]=0$. $E[X^2]=1$ and $E[Q^2]=4$.
$E[ZY] = \frac{4}{5}(1) - \frac{1}{5}(4) = 0$. This confirms the residual is completely uncorrelated with the observation $Y$.

### 5. Information Theory: Differential Entropy
* $h(X) = \frac{1}{2} \ln(2\pi e \sigma_X^2) = \frac{1}{2} \ln(2\pi e (1)) \approx 1.4189$ nats.
* $\text{Var}(X \mid Y) = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = \frac{4}{5} = 0.8$.
* $h(X \mid Y) = \frac{1}{2} \ln(2\pi e \sigma_{X|Y}^2) = \frac{1}{2} \ln(2\pi e (0.8)) \approx 1.3073$ nats.

### 6. Mutual Information
* $I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \ln(2\pi e) - \frac{1}{2} \ln(2\pi e \cdot 0.8) = -\frac{1}{2} \ln(0.8) \approx 0.1116$ nats.
* If variance of $Q$ becomes $100$: $\text{Var}(Y) = 101$. $\text{Var}(X \mid Y) = 1 - \frac{1}{101} = \frac{100}{101}$.
  $I(X; Y) = -\frac{1}{2} \ln(\frac{100}{101}) = \frac{1}{2} \ln(1.01) \approx 0.0049$ nats. The information gained becomes almost zero due to immense noise.

### 7. Diagnostic Interpretation
If we use $\widehat{X}_{naive} = Y$, then conditioned on a specific observation $y$, the expected true temperature is $E[X \mid Y=y] = y/5$. Using $Y$ directly systematically overestimates the magnitude of the fault. Because the variance of the noise ($Q$) is 4x larger than the signal ($X$), extreme values of $Y$ are probabilistically much more likely to be driven by noise. $E[X \mid Y]$ acts as a Bayes estimator, "shrinking" the prediction toward the prior mean (0) proportional to the unreliability of the measurement.

```python
import numpy as np

# Variances and Covariances
var_X = 1
var_Q = 4
var_Y = var_X + var_Q
cov_XY = var_X

# Verification of Orthogonality
E_X2 = var_X
E_Q2 = var_Q
E_ZY = (4/5) * E_X2 - (1/5) * E_Q2
print(f"E[ZY] (Orthogonality check): {E_ZY}")

# Differential Entropy (nats)
h_X = 0.5 * np.log(2 * np.pi * np.e * var_X)

var_X_given_Y = var_X - (cov_XY**2 / var_Y)
h_X_given_Y = 0.5 * np.log(2 * np.pi * np.e * var_X_given_Y)

I_XY = h_X - h_X_given_Y

print(f"h(X): {h_X:.4f} nats")
print(f"h(X|Y): {h_X_given_Y:.4f} nats")
print(f"I(X;Y) with var(Q)=4: {I_XY:.4f} nats")

# Change variance of Q to 100
var_Q_new = 100
var_Y_new = var_X + var_Q_new
var_X_given_Y_new = var_X - (cov_XY**2 / var_Y_new)
I_XY_new = 0.5 * np.log(1) - 0.5 * np.log(var_X_given_Y_new)
print(f"I(X;Y) with var(Q)=100: {I_XY_new:.4f} nats")
```
