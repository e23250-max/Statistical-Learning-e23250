# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

1.  The canonical partition function is defined as:$$Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$$Substitute the given energy eigenvalues $E_j = \hbar \omega (j + \frac{1}{2})$:$$Z(T) = \sum_{j=0}^\infty \exp\left(-\frac{\hbar \omega (j + 1/2)}{k_B T}\right)$$$$Z(T) = \exp\left(-\frac{\hbar \omega}{2k_B T}\right) \sum_{j=0}^\infty \exp\left(-\frac{\hbar \omega j}{k_B T}\right)$$The summation is a geometric series of the form $\sum_{j=0}^\infty r^j = \frac{1}{1-r}$, where $r = e^{-\frac{\hbar \omega}{k_B T}}$. Because $\hbar \omega > 0$ and $T > 0$, we know $r < 1$, guaranteeing convergence.Applying the geometric series sum formula yields:$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

2.  Helmholtz Free Energy $F(T)$:$$F(T) = -k_B T \ln Z(T)$$$$F(T) = -k_B T \left[ \ln\left(e^{-\frac{\hbar \omega}{2k_B T}}\right) - \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right]$$$$F(T) = -k_B T \left( -\frac{\hbar \omega}{2k_B T} \right) + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$$$$F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$$Internal Energy $U(T)$:Let $\beta = \frac{1}{k_B T}$. The partition function is $Z = \frac{e^{-\beta \hbar \omega / 2}}{1 - e^{-\beta \hbar \omega}}$.$$U(T) = -\frac{\partial \ln Z}{\partial \beta} = -\frac{\partial}{\partial \beta} \left[ -\frac{\beta \hbar \omega}{2} - \ln\left(1 - e^{-\beta \hbar \omega}\right) \right]$$$$U(T) = -\left[ -\frac{\hbar \omega}{2} - \frac{1}{1 - e^{-\beta \hbar \omega}} (\hbar \omega e^{-\beta \hbar \omega}) \right]$$$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1 - e^{-\beta \hbar \omega}}$$Multiply numerator and denominator of the second term by $e^{\beta \hbar \omega}$:$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$$

3. Using the relation $S = -\left(\frac{\partial F}{\partial T}\right)$:$$S(T) = -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right]$$$$S(T) = -\left[ k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \left( \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \right) \left( -e^{-\frac{\hbar \omega}{k_B T}} \right) \left( \frac{\hbar \omega}{k_B T^2} \right) \right]$$$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_B T}} - 1}$$Verification ($U = F + TS$):$$F + TS = \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) \right] + T \left[ -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_B T}} - 1} \right]$$The $k_B T \ln(\dots)$ terms cancel out, leaving:$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1} = U(T)$$

4.  $$C_V = \left(\frac{\partial U}{\partial T}\right) = \frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + \hbar \omega \left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-1} \right]$$$$C_V = \hbar \omega \left[ -\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-2} \cdot e^{\frac{\hbar \omega}{k_B T}} \cdot \left(-\frac{\hbar \omega}{k_B T^2}\right) \right]$$$$C_V = k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 \frac{e^{\frac{\hbar \omega}{k_B T}}}{\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^2}$$Rewrite using the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$.Divide the numerator and denominator by $e^{\frac{\hbar \omega}{k_B T}}$:$$C_V = k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 \frac{1}{e^{\frac{\hbar \omega}{k_B T}} - 2 + e^{-\frac{\hbar \omega}{k_B T}}}$$Note that the denominator is a perfect square: $\left(e^{\frac{\hbar \omega}{2k_B T}} - e^{-\frac{\hbar \omega}{2k_B T}}\right)^2$.Substitute $x = \frac{\hbar \omega}{2k_B T}$. We know $(2\sinh(x))^2 = 4\sinh^2(x) = (e^x - e^{-x})^2$.$$C_V = k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 \frac{1}{4 \sinh^2\left(\frac{\hbar \omega}{2k_B T}\right)} = k_B \left(\frac{\hbar \omega}{2k_B T}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_B T}\right)}$$

5.  In the high-temperature limit, $k_B T \gg \hbar \omega$, the argument $x = \frac{\hbar \omega}{2k_B T}$ approaches 0.For small $x$, the Taylor expansion gives $\sinh(x) \approx x$.$$C_V \approx k_B (x)^2 \frac{1}{(x)^2} = k_B$$Does it match? Yes. According to the classical equipartition theorem, a 1-dimensional harmonic oscillator has two degrees of freedom (one for kinetic energy, one for potential energy). Each degree of freedom contributes $\frac{1}{2} k_B$ to the heat capacity, resulting in a total classical prediction of $C_V = k_B$. The quantum model correctly recovers this classical limit at high temperatures.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

We want to find the probability density function $p(x)$ that maximizes the differential entropy:$$H(X) = -\int_{-\infty}^{\infty} p(x) \ln p(x) dx$$Subject to three constraints:Normalization: $\int p(x) dx = 1$Mean: $\int x p(x) dx = \mu$Variance: $\int (x - \mu)^2 p(x) dx = \sigma^2$We use the method of Lagrange multipliers. The functional to maximize is:$$\mathcal{L}[p] = -\int p(x) \ln p(x) dx + \lambda_0 \left(1 - \int p(x) dx\right) + \lambda_1 \left(\mu - \int x p(x) dx\right) + \lambda_2 \left(\sigma^2 - \int (x - \mu)^2 p(x) dx\right)$$Taking the functional derivative with respect to $p(x)$ and setting it to zero (Euler-Lagrange equation):$$\frac{\delta \mathcal{L}}{\delta p(x)} = -1 - \ln p(x) - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$$$\ln p(x) = -1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2$$$$p(x) = \exp\left(-1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2\right)$$To ensure the distribution is symmetric around the mean $\mu$ (as required by the variance constraint without higher-order skewed moments), the linear multiplier $\lambda_1$ must equal 0.Let $C = e^{-1 - \lambda_0}$, then:$$p(x) = C \exp\left(-\lambda_2 (x - \mu)^2\right)$$For the integral to converge, $\lambda_2 > 0$. We use the standard Gaussian integrals to solve for the constants:From Constraint 1 (Normalization):$$C \int_{-\infty}^{\infty} \exp\left(-\lambda_2 (x - \mu)^2\right) dx = 1 \implies C \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies C = \sqrt{\frac{\lambda_2}{\pi}}$$From Constraint 3 (Variance):$$C \int_{-\infty}^{\infty} (x - \mu)^2 \exp\left(-\lambda_2 (x - \mu)^2\right) dx = \sigma^2$$$$\sqrt{\frac{\lambda_2}{\pi}} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2 \implies \frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$Substitute $\lambda_2$ back into $C$:$$C = \sqrt{\frac{1 / (2\sigma^2)}{\pi}} = \frac{1}{\sqrt{2\pi\sigma^2}}$$Therefore, the maximum entropy distribution is the normal distribution:$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




By the Maximum Entropy principle for conditional distributions, subject to feature constraints of the form $\mathbb{E}[f_i(x, Y)] = c_i$, the optimal probability distribution takes the exponential family form (which is identical to Logistic Regression for binary classification):$$P(Y=1 | \mathbf{x}) = \frac{\exp(w_1 x_1 + w_2 x_2 + w_3 x_3)}{1 + \exp(w_1 x_1 + w_2 x_2 + w_3 x_3)} = \sigma(\mathbf{w}^T \mathbf{x})$$where $\sigma(\cdot)$ is the sigmoid function, and $w_1, w_2, w_3$ are the Lagrange multipliers associated with the three constraints.Let $p_j = P(Y=1 | \mathbf{x}_j)$ for scenarios $j \in \{A, B, C, D, E\}$. We can express the constraints as a system of equations based on the marginal probabilities of the scenarios:Constraint 1 ($x_1$): Scenarios A and E have $x_1 = 1$.$$P(\mathbf{x}_A) p_A + P(\mathbf{x}_E) p_E = 0.05 \implies 0.40 p_A + 0.05 p_E = 0.05$$Constraint 2 ($x_2$): Scenarios B, D, and E have $x_2 = 1$.$$P(\mathbf{x}_B) p_B + P(\mathbf{x}_D) p_D + P(\mathbf{x}_E) p_E = 0.15 \implies 0.25 p_B + 0.15 p_D + 0.05 p_E = 0.15$$Constraint 3 ($x_3$): Scenarios C and D have $x_3 = 1$.$$P(\mathbf{x}_C) p_C + P(\mathbf{x}_D) p_D = 0.16 \implies 0.05 p_C + 0.15 p_D = 0.16$$Using the MaxEnt formulation, the probabilities are tied by the weights:$p_A = \sigma(w_1)$$p_B = \sigma(w_2)$$p_C = \sigma(w_3)$$p_D = \sigma(w_2 + w_3)$$p_E = \sigma(w_1 + w_2)$This creates a strictly convex optimization problem. Solving this system of equations numerically yields the specific weights $\mathbf{w}$ that calibrate the AI routing agent, perfectly matching the empirical distribution from the historical audits while making no additional assumptions about the overlapping scenarios.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



$$H(S_T) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) dS_T$$Subject to Market Constraints:$\int_{0}^{\infty} p(S_T) dS_T = 1$ (Probabilities sum to 1)$\int_{0}^{\infty} S_T p(S_T) dS_T = 102$ (Risk-Free Forward Price)$\int_{0}^{\infty} S_T^2 p(S_T) dS_T = 10600$ (Market Volatility)$\int_{0}^{\infty} \ln(S_T) p(S_T) dS_T = 4.60$ (Asymmetric Skew/Crash Risk)Formulation:We define the Lagrangian with multipliers $\lambda_0, \lambda_1, \lambda_2, \lambda_3$:$$\mathcal{L} = -\int p(S_T) \ln p(S_T) dS_T + \lambda_0 \left(1 - \int p(S_T) dS_T\right) + \lambda_1 \left(102 - \int S_T p(S_T) dS_T\right) + \lambda_2 \left(10600 - \int S_T^2 p(S_T) dS_T\right) + \lambda_3 \left(4.60 - \int \ln(S_T) p(S_T) dS_T\right)$$Applying the calculus of variations, we take the derivative with respect to $p(S_T)$ and set it to 0:$$\frac{\partial \mathcal{L}}{\partial p(S_T)} = -1 - \ln p(S_T) - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$Solving for $p(S_T)$:$$\ln p(S_T) = -1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)$$$$p(S_T) = \exp\left(-1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)\right)$$Using the property of logarithms $e^{-\lambda_3 \ln(S_T)} = S_T^{-\lambda_3}$, we can separate the constants. Let $Z = e^{1+\lambda_0}$ be the partition function (normalizing constant):$$p(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right) \quad \text{for } S_T > 0$$
